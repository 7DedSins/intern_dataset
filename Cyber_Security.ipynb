```
df.drop(['req/baseUrl', 'req/fresh', 'req/headers/host', 'req/headers/user-agent', 
       'req/headers/content-type', 'req/headers/org_id', 
       'req/headers/user_session_id', 'req/headers/accept',
       'req/headers/content-length', 'req/headers/user/name',
       'req/headers/user/role', 'req/hostname', 'req/ip', 'req/originalUrl',
       'req/path', 'req/protocol', 'req/secure', 'req/stale',
       'req/subdomains/0', 'req/xhr'], inplace = True, axis = 1)

df.head()
```

Dropping all other columns other than 'isSafe', 'desc' and 'title' columns. Because, only these three columns have unique elements. All the other columns have only one element in the entire dataset. As such, these columns are nothing but overfitting of the data that will overfit all the other models used on the dataset. As such, the other columns and their elements are dropped to yield better results.

```
df["isSafe"] = df["isSafe"].astype(int)
df.head()
```

'isSafe' column contains the boolean values in the dataset. These are treated as strings during model training. Thus, converting the boolean into integer will yeild better result during training of dataset.

```
from sklearn import preprocessing 

label_encoder = preprocessing.LabelEncoder() 

df['req/body/note/title']= label_encoder.fit_transform(df['req/body/note/title']) 

df.head()
```

'title' column has unique elements. Even though this column contains names, yet some are unique out of 1000. Thus, label encoding is used to help train the dataset. One Hot Encoding is not used because it would cause too many unnecessary columns to be created and thus, might overfit the data during training.

```
for i in range(1000):
  df['req/body/note/desc'][i] = len(df['req/body/note/desc'][i])
```

'desc' column has texts present in it with many unique features. However, all are strings that cannot be used during model training. So, the texts are converted to length of the text, a numerical feature in order to aid the training.

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```
y_pred = model.predict(x_test)
```

```
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
mae=mean_absolute_error(y_test, y_pred)
mse=mean_squared_error(y_test, y_pred)
rmse=np.sqrt(mse)
r2_score=r2_score(y_test, y_pred)
print(r2_score)
```

This clearly shows that the Linear Regression model yielded very poor results, with an accuracy of 29%. Even though, the accuracy yielded was quite good given the dataset with no unique features to calculate any productive result.
